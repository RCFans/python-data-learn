```python
import pandas as pd
import time
chunkSize = 50000000
location = 'data/servicelog'
reader = pd.read_csv(location, header=None, sep='\t', iterator=True)
```

```python
print chunkSize
start = time.time()
loop = True
chunks = []
while loop:
    try:
        chunk = reader.get_chunk(chunkSize)
        chunks.append(chunk)
    except StopIteration:
        loop = False
        print "Iteration is stopped."
print time.time() - start
df = pd.concat(chunks, ignore_index=True)
print time.time() - start
```
